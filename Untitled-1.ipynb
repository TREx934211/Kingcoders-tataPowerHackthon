```python
from __future__ import absolute_import, division, print_function, unicode_literals

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output
from six.moves import urllib

import tensorflow.compat.v2.feature_column as fc

import tensorflow as tf

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler
from tensorflow.keras.regularizers import l2
```

```python
input_data = pd.read_csv("Data/spg.csv")

input_data = input_data.astype(int)


test_size_ratio = 0.2

num_column = input_data.shape[0]
num_test = int(test_size_ratio * num_column )
input_data = input_data.sample(frac=1, random_state=32)

train_data = input_data[num_test:]
test_data = input_data[:num_test]

y_train = train_data.pop('generated_power_kw')
y_test = test_data.pop('generated_power_kw')

test_data
```

```python
model = Sequential([
    Dense(1000, activation='relu', input_shape=(20,), kernel_regularizer=l2(0.001)),
    Dense(500, activation='relu', kernel_regularizer=l2(0.001)),
    Dense(1, activation='linear', kernel_regularizer=l2(0.0001))  # Output layer with 1 neuron for regression
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.000255), loss='mse', metrics=['accuracy'])
```

```python
model.fit(train_data, y_train, epochs=100)
```

```python
model.summary()
```

```python
loss = model.evaluate(test_data, y_test)
print("Mean Squared Error on Test Data:", loss)
```
